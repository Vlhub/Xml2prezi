## Перевод

Определенный интерес представляет расширение «nbTranslate», которое  помогает конвертировать текст с одного языка на другой. Расширение обеспечивает многоязычную поддержку переводов. Это расширение автоматически копирует исходную ячейку в новую, но уже на другом языке, а далее предоставляет возможность для редактирования перевода. Базовые структуры текста (например, полужирный, курсив, списки, рисунки, формулы) сохраняются  после преобразования. 

Метаданные, указывающие используемый язык, добавляются в каждый фрагмент текста. Это позволяет выборочно отображать ячейки для определенного языка и скрывать другие. Кодовые ячейки остаются нетронутыми. Таким образом, можно получить своего рода многоязычную научную статью. Настройки для выбора языков доступны через соответствующую кнопку на верхней панели инструментов.

Для примера создадим фрагмент текста предназначенного для перевода на английский язык. Для этого предварительно настроим первичный (языка с которого осуществляется перевод) и вторичный (язык на который осуществляется перевод) с помощью конфигуратора «nbTranslate». Воспользуемся кнопкой панели инструментов "Translate current cell" и в результате получим копию данной ячейки, но на другом (выбранном) языке.

For example, create a piece of text intended for translation into English. To do this, first configure the primary (the language from which the translation is made) and the secondary (the language into which the translation is made) using the “nbTranslate” configurator. We use the button of the toolbar "Translate current cell" and as a result we will receive a copy of this cell, but in a different (selected) language.

В метаданных фрагмента на русском языке будет установлено значение словаря:

```
{
  "lang": "ru"
}
```

В метаданных фрагмента текста на английском языке будет соответственно:

```
{
  "lang": "en"
}

Чтобы оставить отображение фрагментов только для определенного языка и скрыть другие достаточно воспользоваться пунктом меню "Langs", где выбрать для какого языка отображать фрагменты текста.

## Выделение текста
C помощью Markdown вы можете использовать три вида выделения: курсив, полужирный или полужирный курсив. Достичь этого можно, окружив необходимый для выделения текст звездочками, например:

* *курсив* = \*курсив\*
* **полужирный**  = \*\*полужирный\*\* или двойной символ подчеркивания \_\_*полужирный\_\_
* ***полужирный курсив*** = полужирный курсив или три символа подчеркивания
* ~~Зачеркнутый~~ = \~\~Зачеркнутый\~\~

Выделение можно применять к отдельному слову, нескольким словам или части слова:
* к **одному** слову  к одному слову 
* ***к нескольким словам***  к нескольким словам 
* **части**чно к части слова

### Ссылки на web-ресурс

Есть два способа создать ссылку на какой-либо web-ресурс: 
* Первый - поместить ссылку прямо в текст. Заключите текст, который пользователь должен видеть и который будет выглядеть как ссылка в квадратные скобки, а URL страницы в круглые так, чтобы комплекты скобок были написаны слитно. Также вы можете добавить дополнительное описание, заключив его в кавычки и поместив после URL.

```Это [пример ссылки 1](http://example.com/ "Всплывающая подсказка") помещенной в предложение.```

Это [пример ссылки 1](http://example.com/ "Всплывающая подсказка") помещенной в предложение.

* Второй способ удобен для длинных ссылок. Чтобы не загромождать текст, вы можете вставлять их подобно сноскам. Добавьте короткий идентификатор с помощью второй пары квадратных скобок (идентификатор может быть поясняющим словом, фразой или числом).

```Это [пример ссылки 2][пс2] помещенной в предложение.```

Это [пример ссылки 2][пс2] помещенной в предложение.

Затем, в любом месте документа (предпочтительней сразу после абзаца или вместе с другими ссылками в конце документа) вы можете указать URL, связав его с идентификатором, например:

[пс]: http://example.com/ "Дополнительное описание"

Результат второго способа будет точно такой же, как и у первого. Это просто организационно более правильный способ создания документа, в будущем упрощающий его редактирование (особенно, если вы работаете не один).

* И последнее. Если вы хотите, чтобы ссылка представляла из себя сам URL, то просто поместите его в угловые скобки: Например: посетите <http://moodle.org/> сейчас!

Для оформления кода в ячейке вида Markdown можно использовать несколько способов:
